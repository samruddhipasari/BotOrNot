```python
gen_users = pd.read_csv('/Users/atulyasharma/Documents/M.S in Information Systems/University Curriculum/Spring 2020/INFO 6105 Data Science Engineering Methods and Tools/Projects/Dataset/cresci-2017.csv/datasets_full.csv/genuine_accounts.csv/users.csv')
gen_users.head()
```

```python
gen_tweets = pd.read_csv('/Users/atulyasharma/Documents/M.S in Information Systems/University Curriculum/Spring 2020/INFO 6105 Data Science Engineering Methods and Tools/Projects/Dataset/cresci-2017.csv/datasets_full.csv/genuine_accounts.csv/tweets.csv')
gen_tweets.head(2839362)
```

```python
gen_tweets.dtypes
```

```python
print(gen_users.shape)
print(gen_tweets.shape)
```

```python
gen_users = gen_users.fillna(0)
gen_tweets = gen_tweets.fillna(0)
```

```python
gen_users['id'].unique().shape
```

```python
gen_tweets['user_id'].unique().shape
```

```python
rem_col = ['default_profile', 'default_profile_image', 'geo_enabled', 'profile_image_url', 'profile_banner_url',
           'profile_use_background_image', 'profile_background_image_url_https', 'profile_text_color',
           'profile_sidebar_border_color', 'profile_background_tile', 'profile_sidebar_fill_color',
           'profile_background_image_url','profile_background_color', 'profile_link_color', 'utc_offset',
           'is_translator', 'protected', 'description', 'contributors_enabled' ]
gen_users = gen_users.drop(rem_col, axis = 1)

rem_col = ['url', 'time_zone', 'location', 'profile_image_url_https']
gen_users = gen_users.drop(rem_col, axis = 1)

rem_col = ['screen_name', 'listed_count', 'follow_request_sent', 'notifications']
gen_users = gen_users.drop(rem_col, axis = 1)
```

```python
new_features = ['in_reply_to_user_id','retweet_count','num_hashtags', 'num_urls','num_mentions']

for f in new_features:
    gen_users[f] = np.nan
```

```python
gen_users['retweet_count'] = gen_users.retweet_count.astype('float64')
gen_users['num_hashtags'] = gen_users.num_hashtags.astype('float64')
gen_users['num_urls'] = gen_users.num_urls.astype('float64')
gen_users['statuses_count'] = gen_users.statuses_count.astype('float64')
gen_users['followers_count'] = gen_users.followers_count.astype('float64')
gen_users['friends_count'] = gen_users.friends_count.astype('float64')
gen_users['favourites_count'] = gen_users.favourites_count.astype('float64')
```

```python
gen_tweets = gen_tweets.drop(2839361)
```

```python
gen_users.shape
```

```python
gen_tweets["id"] = pd.to_numeric(gen_tweets["id"])
```

```python
unique_ids = gen_tweets['user_id'].unique()
indices = []

for unique_id in unique_ids:
    index = gen_users.loc[gen_users['id']==unique_id].index[0]
    indices.append(gen_users.loc[gen_users['id']==unique_id].index[0])
    
    mean = gen_tweets.loc[gen_tweets['user_id'] == unique_id]['num_hashtags'].mean()
    gen_users.iloc[index, gen_users.columns.get_loc('num_hashtags')] = mean
    
    mean = gen_tweets.loc[gen_tweets['user_id'] == unique_id]['in_reply_to_user_id'].mean()
    gen_users.iloc[index, gen_users.columns.get_loc('in_reply_to_user_id')] = mean

    mean = gen_tweets.loc[gen_tweets['user_id'] == unique_id]['retweet_count'].mean()
    gen_users.iloc[index, gen_users.columns.get_loc('retweet_count')] = mean

    mean = gen_tweets.loc[gen_tweets['user_id'] == unique_id]['num_urls'].mean()
    gen_users.iloc[index, gen_users.columns.get_loc('num_urls')] = mean

    mean = gen_tweets.loc[gen_tweets['user_id'] == unique_id]['num_mentions'].mean()
    gen_users.iloc[index, gen_users.columns.get_loc('num_mentions')] = mean
```

```python
gen_users.dtypes
```

```python
age = (pd.to_datetime('2018-11-09 00:00:00') - pd.to_datetime(gen_users['timestamp']))/np.timedelta64(1, 'M')
gen_users.loc[:,'acct_age'] = age.values
gen_users = gen_users.drop('timestamp',1)
```

```python
gen_users = gen_users.dropna(how='any')
```

```python
gen_users.dtypes
```

```python
rem_col = ['created_at', 'crawled_at', 'updated', 'test_set_1', 'test_set_2']
gen_users = gen_users.drop(rem_col, axis = 1)
```

```python
features = ['statuses_count', 'followers_count', 'friends_count', 'favourites_count', 'verified', 'following', 'in_reply_to_user_id', 
            'retweet_count', 'num_hashtags', 'num_urls', 'num_mentions', 'acct_age']

for feature in features:
    max_value = np.maximum(gen_users[feature].max(), gen_users[feature].max())
    gen_users[feature] = gen_users[feature]/max_value
```

```python
gen_users['user_type'] = 1
```

```python
gen_users.head()
```

```python
gen_users = gen_users.fillna(0)
```

```python
gen_users.lang.unique()
```

```python
gen_users['language'] = gen_users['lang'].apply(lambda x: x.split('-')[0])
del gen_users['lang']
gen_users.head()
```

```python
gen_users.rename(columns = {'language' : 'lang'}, inplace = True)
```

```python
gen_users.isnull().sum()
```

```python
gen_users.to_csv('gen_data_final.csv', sep=',')
```
